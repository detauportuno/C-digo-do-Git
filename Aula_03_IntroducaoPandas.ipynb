# Corrigindo erros do Dataset

```python
df.columns
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df.head()
```

```python
def transforma_texto_para_inteiro(valor):
  multiplicadores = {
      "K": 1000,
      "M": 1000000
  }
  valor_str = valor.upper()

  for suffix, multiplicador in multiplicadores.items():
    if valor_str.endswith(suffix):
      numeros = valor_str[:-1]
      numeros = int(float(numeros) * multiplicador)
      return numeros
  return int(valor_str)
  # print(valor_str)
```

```python
transforma_texto_para_inteiro("15")
```

```python
df['Plays'] = df['Plays'].apply(transforma_texto_para_inteiro)
df['Times Listed'] = df['Times Listed'].apply(transforma_texto_para_inteiro)
df['Number of Reviews'] = df['Number of Reviews'].apply(transforma_texto_para_inteiro)
df['Playing'] = df['Playing'].apply(transforma_texto_para_inteiro)
df['Backlogs'] = df['Backlogs'].apply(transforma_texto_para_inteiro)
df['Wishlist'] = df['Wishlist'].apply(transforma_texto_para_inteiro)
```
